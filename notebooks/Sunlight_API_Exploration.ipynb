```python
officials_list = []

for i,zipcode in enumerate(zip_list):
    try:
        officials = votesmart.officials.getByZip(zipcode)
        for j,official in enumerate(officials):
            new_record = {
                'zipcode' : zipcode,
                'zip_latitude' : zip_lat_list[i],
                'zip_longitude' : zip_lng_list[i],
                'candidateId' : officials[j].candidateId,
                'firstName' : officials[j].firstName,
                'middleName':  officials[j].middleName,
                'preferredName' : officials[j].preferredName,
                'lastName' : officials[j].lastName,
                'suffix' : officials[j].suffix,
                'nickName' : officials[j].nickName,
                'title' : officials[j].title,
                'officeParties' : officials[j].officeParties,
                'officeStatus' : officials[j].officeStatus,
                'officeDistrictId' : officials[j].officeDistrictId,
                'officeDistrictName' : officials[j].officeDistrictName,
                'officeStateId' : officials[j].officeStateId,
                'officeId' : officials[j].officeId,
                'officeName' : officials[j].officeName,
                'officeTypeId' : officials[j].officeTypeId
            }

            officials_list.append(new_record)
    except:
        pass

officials_df = pd.DataFrame(officials_list)
    
```
