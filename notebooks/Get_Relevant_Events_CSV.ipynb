```python
indivisible_groups = []


queryURL = 'https://www.indivisibleguide.com/groups-nav'


display = Display(visible=0, size=(800, 600))
display.start()
browser = webdriver.Firefox()
browser.get(queryURL)
# browser.implicitly_wait(30)
# browser.find_element_by_xpath('//*[@id="map-container"]/div[1]/div[2]/div[2]/svg/g[4100]/path').click()

for i in range(860):
    for j in range(758):
        
        x = i # set actual value
        y = j # set actual value
        length = 860 # set actual value
        width = 758 # set actual value
        
        if width - y > 72 and width - y < 750 and length - x > 350 and length - x < 770:
            WebDriverWait(browser, 30).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="map-container"]')))
            action = webdriver.common.action_chains.ActionChains(browser)
            element=browser.find_elements_by_xpath('//*[@id="map-container"]') # specify correct xpath
            action.move_to_element_with_offset(element, width - y, length - x)
            action.click()
            action.perform()
            time.sleep(3)
            html = browser.page_source
            soup = soupy(html)

            divs = soup.findAll("div", { "class" : "event-item lato group" })


            for div in divs:
                lat = div['lat']
                lng = div['lon']
                group = div.h3
                Links = [dv.a['href'] for dv in div.findAll("div", { "class" : "event-social" })]

                new_record ={
                    'latitude':lat,
                    'longitude':lng,
                    'group_name':group,
                    'group_links':Links
                }
                indivisible_groups.append(new_record)

# <div class="event-item lato group" lat="39.0882" lon="-79.4563">
#               <h5 class="time-info">
#                 <span class="time-info-dist"></span>
#               </h5>
#               <h3>
#                 <a target="_blank" href="https://www.facebook.com/wvhighlandsindivisible/">WV Highlands Indivisible</a>
#               </h3>
#               <span class="label-icon"></span>
#               <h5 class="event-type">Group</h5>
#               <div class="event-social">
#                 <a href="https://www.facebook.com/wvhighlandsindivisible/" target="_blank"><img src="/img/icon/facebook.png"></a><a href="mailto:skymist@frontier.com"><img src="/img/icon/mailchimp.png"></a>
#               </div>
#             </div>



time.sleep(5) # pause/sleeps for 5 seconds
browser.quit()
display.stop()
```
